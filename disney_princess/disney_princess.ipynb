```python
top_10 = data\
    .sort_values(by='PopularityScore', ascending=False)\
    .drop_duplicates(subset='PrincessName', keep='first')

top_10
```

```python
plt.figure(figsize = (10,6))
sns.histplot(data['MovieRuntimeMinutes'], kde=True, color='skyblue')
plt.title('Movie Runtime')
```

```python
unique_princess = data.drop_duplicates(subset='PrincessName')
unique_princess.head()

plt.figure(figsize = (10,6))
sns.countplot(y='HairColor', data=unique_princess, order=unique_princess['HairColor'].value_counts().index, hue='HairColor', palette='pastel', legend=False)
hair_color_groups = unique_princess.groupby('HairColor')['PrincessName'].apply(list)

# Print nicely
for color, princesses in hair_color_groups.items():
    print(f"{color}: {princesses}")

```
