```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('./data/spotify_top_1000_tracks.csv')
```

```python
df[['year', 'month', 'day']] = df['release_date'].str.split('-', expand=True)
```

```python
year_2024.sort_values(by=['popularity'], ascending=True)
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x=df['year'], y=df['popularity'])
plt.title('Popularity Over Year',fontsize=15)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Popularity', fontsize=12)

plt.grid()
plt.show()
```
