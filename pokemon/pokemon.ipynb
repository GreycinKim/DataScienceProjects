```python
df['type2'] = df['type2'].fillna('None')
df['percentage_male'] = df['percentage_male'].fillna(-1)
df['height_m'] = df['height_m'].fillna(df['height_m'].median())
df['weight_kg'] = df['weight_kg'].fillna(df['weight_kg'].median())
```

```python
def top5_by_type(pokemon_df, type_name):
    filtered = pokemon_df[pokemon_df['type1'] == type_name]
    top5 = filtered.sort_values(by='base_total', ascending=True).head(5)
    return top5[['name', 'base_total', 'attack', 'defense', 'speed']]
```

```python
top5_by_type(df, 'fire')
```

```python
top5_by_type(df, 'water')
```

```python
top5_by_type(df, 'grass')
```
